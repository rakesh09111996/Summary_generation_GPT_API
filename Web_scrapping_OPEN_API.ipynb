```python
text_extract = scrap.get_text()
print(text_extract)
```

```python
from openai import OpenAI

client = OpenAI(api_key = '$OPEN_PI_KEY')

stream = client.chat.completions.create(
    model="gpt-4",
    messages=[{ "role": "system", "content": "you are summarizer, given a text summarise the important points alone and ignore unnecessary details"},
        {"role": "user", "content": text_extract}],
)
print(stream.choices[0].message.content)
```
